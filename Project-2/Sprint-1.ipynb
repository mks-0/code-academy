```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Import the CSVs
data_1 = pd.read_csv('data/winequality-red.csv', sep=';')
data_2 = pd.read_csv('data/winequality-white.csv', sep=';')
# Make copies of the dataframes
rw = data_1.copy()
ww = data_2.copy()
```

```python
# Creating a new column called "quality label"
rw["quality label"] = rw["quality"].apply(
    lambda value: "low" if value <= 5 else "medium" if value < 7 else "high"
)
ww["quality label"] = ww["quality"].apply(
    lambda value: "low" if value <= 5 else "medium" if value < 7 else "high"
)

# Transforming these labels into categrical data type
categories=["low", "medium", "high"]
rw["quality label"] = pd.Categorical(
    rw["quality label"], categories=categories
)
ww["quality label"] = pd.Categorical(
    ww["quality label"], categories=categories
)
```

```python
rw['wine_type'] = 'Red'
ww['wine_type'] = 'White'
rw['wine_type'] = pd.Categorical(rw['wine_type'], categories=['Red'])
ww['wine_type'] = pd.Categorical(ww['wine_type'], categories=['White'])
print(rw.shape[0] + ww.shape[0])
wine = pd.merge(rw, ww, how='outer')
print(wine.shape)
display(wine.head(5))
display(wine.tail(5))
```

```python
cols = [
    "residual sugar",
    "total sulfur dioxide",
    "sulphates",
    "alcohol",
    "volatile acidity",
    "quality",
]
pd.concat(
    [rw[cols].describe(), ww[cols].describe()],
    axis=1,
    keys=["Red Wines Stats", "White Wines Stats"],
)

# White Wine's Residual Sugar mean value is considerably higher
# Total Sulfur Dioxide mean value of White Wine is 3x higher that Red Wine's
# Volatile Acidity mean value of Red Wine is 2x higher that the White Wine's
```

```python
sns.set_theme()
colors = ['#e60c34', '#fda929', '#19ae1d']

# Plot 1. Wine Type Count

wine["wine_type"].value_counts().plot(
    kind="pie", figsize=(6, 6), textprops={"fontsize": 20}, autopct="%1.0f%%", colors = ['#ffffe0', '#e60c34'],
    wedgeprops={'edgecolor': '#000000', 'linewidth': 2}
)

plt.title("Wine Type Count", fontsize=25)
plt.ylabel("")
```

```python
# Plot 2. Quality Labels Count

fig, axes = plt.subplots(1, 2, sharey=True, figsize=(12, 5))

sns.countplot(data=rw, x='quality label', ax=axes[0], palette=colors)
axes[0].set_title('Red Wine', fontsize=20, pad=20)
axes[1].set_title('White Wine', fontsize=20, pad=20)
sns.countplot(data=ww, x='quality label', ax=axes[1], palette=colors)
[ax.set_ylabel("Count", fontsize=15, labelpad=15) for ax in axes]
[ax.set_xlabel("Quality", fontsize=15, labelpad=15) for ax in axes]
    
```

```python
# Plot 3. Wine Heatmap

fig, axes = plt.subplots(3, 1, figsize=(10, 20))
fig.tight_layout(pad=9)

plt.figure(figsize=(8, 6))
mask = np.triu(np.ones_like(rw.corr(numeric_only=True)))
sns.heatmap(
    wine.corr(numeric_only=True), annot=True, fmt=".2f", cmap="RdYlGn", mask=mask, ax=axes[0],
)
axes[0].set_title("Wine", fontsize=15, pad=10)
sns.heatmap(
    rw.corr(numeric_only=True), annot=True, fmt=".2f", cmap="RdYlGn", mask=mask, ax=axes[1],
)
axes[1].set_title("Red Wine", fontsize=15, pad=10)
sns.heatmap(
    ww.corr(numeric_only=True), annot=True, fmt=".2f", cmap="RdYlGn", mask=mask, ax=axes[2],
)
axes[2].set_title("White Wine", fontsize=15, pad=10)
```

```python
# Plot 4. Wine Quality - Alcohol Content/pH

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

fig.suptitle('Wine Quality - Alcohol Content/pH', fontsize=20)

sns.boxplot(data=wine, x='quality label', y='alcohol', ax=axes[0], palette=colors)
axes[0].set_ylabel("Alcohol %", fontsize=15, labelpad=15)
axes[0].set_xlabel("Quality", fontsize=15, labelpad=20)

sns.boxplot(data=wine, x='quality label', y='pH', ax=axes[1], palette=colors)
axes[1].set_ylabel('pH', fontsize=15, labelpad=15)
axes[1].set_xlabel("Quality", fontsize=15, labelpad=20)

```

```python
# Plot 5. Red Wine Properties to Quality Label Correlation 
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Red Wine Properties to Quality Label Correlation ', fontsize=20)

# Positive Correlation
sns.boxplot(data=rw, y='sulphates', x='quality label', ax=axes[0][0], palette=colors)
sns.boxplot(data=rw, y='citric acid', x='quality label', ax=axes[0][1], palette=colors)
sns.boxplot(data=rw, y='alcohol', x='quality label', ax=axes[0][2], palette=colors)
# Negative Correlation
sns.boxplot(data=rw, y='density', x='quality label', ax=axes[1][0], palette=colors)
sns.boxplot(data=rw, y='volatile acidity', x='quality label', ax=axes[1][1], palette=colors)
sns.boxplot(data=rw, y='total sulfur dioxide', x='quality label', ax=axes[1][2], palette=colors)
```

```python
# Plot 6. White Wine Properties to Quality Label Correlation 

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('White Wine Properties to Quality Label Correlation ', fontsize=20)

# Positive Correlation
sns.boxplot(data=ww, y='pH', x='quality label', ax=axes[0][0], palette=colors)
sns.boxplot(data=ww, y='alcohol', x='quality label', ax=axes[0][1], palette=colors)
# Negative Correlation
sns.boxplot(data=ww, y='volatile acidity', x='quality label', ax=axes[1][0], palette=colors)
sns.boxplot(data=ww, y='density', x='quality label', ax=axes[1][1], palette=colors)
sns.boxplot(data=ww, y='chlorides', x='quality label', ax=axes[1][2], palette=colors)
```
