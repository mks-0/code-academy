I've created a new category called "Quality Label" and divided the samples into three groups: \
LOW - samples with score 0 - 4 \
MEDIUM - samples with score 5 - 7 \
HIGH - samples with score 8 - 10 


```python
# Creating a new column called "quality label"
rw["quality label"] = rw["quality"].apply(
    lambda value: "low" if value < 5 else "medium" if value <= 7 else "high"
)
ww["quality label"] = ww["quality"].apply(
    lambda value: "low" if value < 5 else "medium" if value <= 7 else "high"
)

# Transforming these labels into categorical data type
categories=["low", "medium", "high"]
rw["quality label"] = pd.Categorical(
    rw["quality label"], categories=categories
)
ww["quality label"] = pd.Categorical(
    ww["quality label"], categories=categories
)
```

Function to filter the outliers based on the 1.5 IQR range. 

```python
# Function for filtering outliers 

def filter_outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = 1.5 * (q3 - q1)
    return data[(data > (q1 - iqr)) & (data < (q3 + iqr))]
```

```python
# Plot 4. Univariate Analysis Wine between quality labels
def unvaq (data):
    n, m = 3, 4
    fig, axes = plt.subplots(n, m, figsize=(7 * m, 5 * n))
    fig.suptitle(
        f"Univariate Analysis of {data['wine type'][0]} Wine between Quality Labels", fontsize=30, y=0.92
    )
    k = 0
    for i in range(n):
        k = i * m
        for j in range(m):
            for l in range(len(categories)):
                sns.kdeplot(
                data=data,
                x=filter_outliers(data[data['quality label'] == categories[l]].iloc[:, k+j]),
                ax=axes[i][j],
                color=colors[l],
                fill=True,
                )
            axes[i][j].set_xlabel(f'{data.columns[k+j]}', fontsize=23, color="#6b6b6b")
            axes[i][j].set(yticklabels=[])
            axes[i][j].set(ylabel=None)

    axes[0][3].legend(labels=['low', 'medium', 'high'], fontsize=18)        
    axes[n-1][m-1].set_visible(False)
    plt.show()

unvaq(rw)
unvaq(ww)
```

```python
# Plot 6. Red Wine Properties to Quality Label Correlation


def boxplot_quality(data, prop_list, type):
    n, m = 2, 3
    fig, axes = plt.subplots(n, m, figsize=(15, 10))
    k = 0
    if type == "box":
        for i in range(n):
            k = i * m
            for j in range(m):
                sns.boxplot(
                    data=data,
                    y=filter_outliers(data[prop_list[k + j]]),
                    x="quality label",
                    ax=axes[i][j],
                    palette=colors,
                    fliersize = 2,
                    boxprops=dict(edgecolor='black')
                )
                for patch in axes[i][j].patches:
                    r, g, b, a = patch.get_facecolor()
                    patch.set_facecolor((r, g, b, .7))
                plt.setp(axes[i][j].lines, zorder=100)
                sns.stripplot(
                    data=data,
                    y=filter_outliers(data[prop_list[k + j]]),
                    x="quality label",
                    ax=axes[i][j],
                    palette=colors,
                    hue='quality label',
                    jitter=0.4,
                    alpha=0.38,
                    size=3
                )
                axes[i][j].legend([], [], frameon=False)       
    # else:
    #     for i in range(n):
    #         k = i * m
    #         for j in range(m):
    #             sns.boxenplot(
    #                 data=data,
    #                 y=filter_outliers(data[prop_list[k + j]]),
    #                 x="quality",
    #                 ax=axes[i][j],
    #                 palette=colors_e,
    #                 flier_kws = {'s': 7},
    #             )
    if data['wine type'][0] == 'Red':
        fig.suptitle("Red Wine Properties to Quality Label Correlation ", fontsize=20)
        # axes[1][2].set_visible(False)
    else:
        fig.suptitle("White Wine Properties to Quality Label Correlation ", fontsize=20)
        # axes[0][2].set_visible(False)
    
    plt.show()


rw_props_list = [
    "sulphates",
    "citric acid",
    "alcohol",
    "density",
    "volatile acidity",
    "chlorides",
]
ww_props_list = [
    "pH",
    "alcohol",
    "free sulfur dioxide",
    "density",
    "chlorides",
    "volatile acidity",
]
boxplot_quality(rw, rw_props_list, "box")
boxplot_quality(ww, ww_props_list, "box")
```

```python
# Plot 7. Bivariate Analysis. KDE and Scatter Plots.

def kde_scatter_plot(data, x, y):
    hue_order = ["high", "medium", "low"]
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    fig.suptitle(
        f"{data['wine type'][0]} Wine \n {x.name.title()} to {y.name.title()}",
        fontsize=20,
    )
    sns.kdeplot(
        data=data,
        y=filter_outliers(y),
        x=filter_outliers(x),
        hue="quality label",
        palette=colors,
        levels=10,
        ax=axes[0],
    )
    sns.scatterplot(
        data=data,
        y=filter_outliers(y),
        x=filter_outliers(x),
        hue="quality label",
        palette=colors_a,
        ax=axes[1],
        marker="s",
        s=25,
        linewidth=0,
    )
    plt.legend([], [], frameon=False)
    plt.show()
```

```python
# Plot 8. Finding a factor 0 < k < 1, that would represent quality based on wine properties.
# The bigger the factor, the better the quality suppose to be.

k_r = (rw["citric acid"] + rw["alcohol"] + rw["sulphates"]) / (rw["density"] + rw["volatile acidity"])
rw["factor"] = k_r / k_r.abs().max()
k_w = (ww["pH"] + ww["alcohol"]) / (
    ww["density"] + ww["volatile acidity"] + ww["chlorides"]
)
ww["factor"] = k_w / k_w.abs().max()

n = 2
fig, axes = plt.subplots(1, n, figsize=(11, 5), sharey=True)
fig.suptitle("Quality Factor", fontsize=20, y=1.1)
q = 'quality'

for i in range(n):
    c = colors_e[:-1] if i == 0 else colors_e
    axes[i].set_title(wine_names[i], fontsize=15)
    sns.boxplot(
        data=data_list[i],
        x=q,
        y="factor",
        palette=c,
        ax=axes[i],
        fliersize=0,
        # linewidth=1,
        
    )
    # for patch in axes[i].patches:
    #     r, g, b, a = patch.get_facecolor()
    #     patch.set_facecolor((r, g, b, .5))
    # plt.setp(axes[i].lines, zorder=100)
    # sns.stripplot(
    #     data=data_list[i],
    #     x=q,
    #     y="factor",
    #     palette=c,
    #     hue=q,
    #     ax=axes[i],
    #     s=2,
    #     jitter=0.4,
    #     alpha=0.5,
    # )
    axes[i].legend([], [], frameon=False)

```

```python
# Testing the probability and ANOVA of the factor

from scipy import stats

def k_prob(data, q):
    q_labels = ["low", "medium", "high"]
    n = q_labels.index(q)
    dql = data["quality label"]
    if n == 0:
        a = data[(data["factor"] < data[dql == q]["factor"].mean()) & (dql == q)]
    else:
        a = data[
            (data["factor"] > data[dql == q_labels[n - 1]]["factor"].mean())
            & (dql == q)
        ]
    return a.value_counts().count() / data[dql == q].value_counts().count()


print(k_prob(rw, "low"), k_prob(rw, "medium"), k_prob(rw, "high"))
print(k_prob(ww, "low"), k_prob(ww, "medium"), k_prob(ww, "high"))
print("\n")

F, p = stats.f_oneway(
    rw[rw["quality label"] == "low"]["factor"],
    rw[rw["quality label"] == "medium"]["factor"],
    rw[rw["quality label"] == "high"]["factor"],
)
print(
    "ANOVA test for mean factor levels across wine samples with different quality ratings"
)
print("F Statistic:", F, "\tp-value:", p)
```

```python
# ANOVA Testing

F, p = stats.f_oneway(
    rw[rw["quality label"] == "low"]["alcohol"],
    rw[rw["quality label"] == "medium"]["alcohol"],
    rw[rw["quality label"] == "high"]["alcohol"],
)
print(
    "ANOVA test for mean alcohol levels across wine samples with different quality ratings"
)
print("F Statistic:", F, "\tp-value:", p)
```
