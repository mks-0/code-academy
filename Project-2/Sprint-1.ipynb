```python
# Creating a new column called "quality label"
rw["quality label"] = rw["quality"].apply(
    lambda value: "low" if value < 5 else "medium" if value < 7 else "high"
)
ww["quality label"] = ww["quality"].apply(
    lambda value: "low" if value < 5 else "medium" if value < 7 else "high"
)

# Transforming these labels into categrical data type
categories=["low", "medium", "high"]
rw["quality label"] = pd.Categorical(
    rw["quality label"], categories=categories
)
ww["quality label"] = pd.Categorical(
    ww["quality label"], categories=categories
)
```

```python
# Creating a DataFrame of both wines

rw['wine_type'] = 'Red'
ww['wine_type'] = 'White'
rw['wine_type'] = pd.Categorical(rw['wine_type'], categories=['Red'])
ww['wine_type'] = pd.Categorical(ww['wine_type'], categories=['White'])
print(rw.shape[0] + ww.shape[0])
wine = pd.merge(rw, ww, how='outer')
print(wine.shape)
display(wine.head(5))
display(wine.tail(5))
```

```python
# Function for filtering outliers 

def filter_outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = 1.5 * (q3 - q1)
    return data[(data > (q1 - iqr)) & (data < (q3 + iqr))]
```

```python
# Plot 5. Red Wine Properties to Quality Label Correlation


def boxplot_quality(data, prop_list, type):
    n, m = 2, 3
    fig, axes = plt.subplots(n, m, figsize=(15, 10))
    k = 0
    if type == "box":
        for i in range(n):
            k = i * m
            for j in range(m):
                sns.boxplot(
                    data=data,
                    y=filter_outliers(data[prop_list[k + j]]),
                    x="quality label",
                    ax=axes[i][j],
                    palette=colors,
                )
    else:
        for i in range(n):
            k = i * m
            for j in range(m):
                sns.boxenplot(
                    data=data,
                    y=filter_outliers(data[prop_list[k + j]]),
                    x="quality",
                    ax=axes[i][j],
                    palette=colors_e,
                )
    if data.equals(rw):
        fig.suptitle("Red Wine Properties to Quality Label Correlation ", fontsize=20)
        axes[1][2].set_visible(False)
    else:
        fig.suptitle("White Wine Properties to Quality Label Correlation ", fontsize=20)
        axes[0][2].set_visible(False)
    plt.show()


rw_props_list = [
    "sulphates",
    "citric acid",
    "alcohol",
    "density",
    "volatile acidity",
    "citric acid",
]
ww_props_list = [
    "pH",
    "alcohol",
    "citric acid",
    "density",
    "chlorides",
    "volatile acidity",
]
boxplot_quality(rw, rw_props_list, "box")
boxplot_quality(rw, rw_props_list, "boxen")
boxplot_quality(ww, ww_props_list, "box")
boxplot_quality(ww, ww_props_list, "boxen")
```

```python
# Plot 6. Univariate Analysis of Red Wine compared to White Wine

n, m = 3, 4
fig, axes = plt.subplots(n, m, figsize=(25, 15))
fig.suptitle(
    "Univariate Analysis of Red Wine compared to White Wine", fontsize=20, y=0.92
)

k = 0

for i in range(n):
    k = i * m
    for j in range(m):
        sns.kdeplot(
        data=rw,
        x=filter_outliers(rw.iloc[:, k+j]),
        ax=axes[i][j],
        color=colors_w[1],
        fill=True,
        cut = 10
        )
        sns.kdeplot(
        data=ww,
        x=filter_outliers(ww.iloc[:, k+j]),
        ax=axes[i][j], 
        color=colors_e[5],
        fill=True
        )
        axes[i][j].set_xlabel(f'{rw.columns[k+j]}', fontsize=16)
        axes[i][j].set(yticklabels=[])
        axes[i][j].set(ylabel=None)
        
axes[2][3].set_visible(False)

```

```python
# Plot 7. Bivariate Analysis. KDE and Scatter Plots.


def kde_scatter_plot(data, x, y):
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    fig.suptitle(f"{x.name.title()} to {y.name.title()}", fontsize=20)
    sns.kdeplot(
        data=data, y=y, x=x, hue="quality label", palette=colors, levels=10, ax=axes[0]
    )
    sns.scatterplot(
        data=data,
        y=y,
        x=x,
        hue="quality label",
        palette=colors_a,
        ax=axes[1],
        marker="s",
        s=30,
        linewidth=0,
    )
    plt.legend([],[], frameon=False)
    plt.show()


kde_scatter_plot(rw, filter_outliers(rw["residual sugar"]), filter_outliers(rw["alcohol"]))
kde_scatter_plot(rw, filter_outliers(rw["volatile acidity"]), filter_outliers(rw["total sulfur dioxide"]))
kde_scatter_plot(rw, filter_outliers(rw["sulphates"]), filter_outliers(rw["alcohol"]))
```

```python
# Plot 8. Finding a factor 0 < k < 1, that would represent quality based on wine properties.
# The bigger the factor, the better the quality suppose to be.

k = (wine["pH"] + wine["alcohol"]) / (
    wine["density"] + wine["volatile acidity"] + wine["chlorides"]
)
wine["factor"] = k / k.abs().max()
k_r = (rw["citric acid"] + rw["alcohol"]) / (rw["density"] + rw["volatile acidity"])
rw["factor"] = k_r / k_r.abs().max()
k_w = (ww["pH"] + ww["alcohol"]) / (
    ww["density"] + ww["volatile acidity"] + ww["chlorides"]
)
ww["factor"] = k_w / k_w.abs().max()

n = 3
fig, axes = plt.subplots(1, n, figsize=(15, 5), sharey=True)
fig.suptitle("Quality Factor", fontsize=20, y=1.1)
q = 'quality'

for i in range(n):
    c = colors_e[:-1] if i == 1 else colors_e
    axes[i].set_title(wine_names[i], fontsize=15)
    sns.stripplot(
        data=data_list[i],
        x=q,
        y="factor",
        palette=c,
        hue=q,
        ax=axes[i],
        s=5,
        jitter=0.45,
        alpha=0.6,
    )
    axes[i].legend([], [], frameon=False)

```

```python
# Testing the probability and ANOVA of the factor
from scipy import stats


def k_prob(data, q):
    q_labels = ["low", "medium", "high"]
    n = q_labels.index(q)
    dql = data["quality label"]
    if n == 0:
        a = data[(data["factor"] < data[dql == q]["factor"].mean()) & (dql == q)]
    else:
        a = data[
            (data["factor"] > data[dql == q_labels[n - 1]]["factor"].mean())
            & (dql == q)
        ]
    return a.value_counts().count() / data[dql == q].value_counts().count()


print(k_prob(rw, "low"), k_prob(rw, "medium"), k_prob(rw, "high"))
print(k_prob(ww, "low"), k_prob(ww, "medium"), k_prob(ww, "high"))
print(k_prob(wine, "low"), k_prob(wine, "medium"), k_prob(wine, "high"))
print("\n")

F, p = stats.f_oneway(
    rw[rw["quality label"] == "low"]["factor"],
    rw[rw["quality label"] == "medium"]["factor"],
    rw[rw["quality label"] == "high"]["factor"],
)
print(
    "ANOVA test for mean factor levels across wine samples with different quality ratings"
)
print("F Statistic:", F, "\tp-value:", p)
```

```python
# ANOVA Testing

F, p = stats.f_oneway(
    rw[rw["quality label"] == "low"]["alcohol"],
    rw[rw["quality label"] == "medium"]["alcohol"],
    rw[rw["quality label"] == "high"]["alcohol"],
)
print(
    "ANOVA test for mean alcohol levels across wine samples with different quality ratings"
)
print("F Statistic:", F, "\tp-value:", p)
```
