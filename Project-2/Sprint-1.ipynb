```python
rw = rw.drop_duplicates()
ww = ww.drop_duplicates()
```

```python
# Creating a new column called "quality label"
rw["quality label"] = rw["quality"].apply(
    lambda value: "low" if value < 5 else "medium" if value < 7 else "high"
)
ww["quality label"] = ww["quality"].apply(
    lambda value: "low" if value < 5 else "medium" if value < 7 else "high"
)

# Transforming these labels into categrical data type
categories=["low", "medium", "high"]
rw["quality label"] = pd.Categorical(
    rw["quality label"], categories=categories
)
ww["quality label"] = pd.Categorical(
    ww["quality label"], categories=categories
)
```

```python
rw['wine_type'] = 'Red'
ww['wine_type'] = 'White'
rw['wine_type'] = pd.Categorical(rw['wine_type'], categories=['Red'])
ww['wine_type'] = pd.Categorical(ww['wine_type'], categories=['White'])
print(rw.shape[0] + ww.shape[0])
wine = pd.merge(rw, ww, how='outer')
print(wine.shape)
display(wine.head(5))
display(wine.tail(5))
```

```python
cols = [
    "residual sugar",
    "total sulfur dioxide",
    "sulphates",
    "alcohol",
    "volatile acidity",
    "quality",
]
pd.concat(
    [rw[cols].describe(), ww[cols].describe()],
    axis=1,
    keys=["Red Wines Stats", "White Wines Stats"],
)

# White Wine's Residual Sugar mean value is considerably higher
# Total Sulfur Dioxide mean value of White Wine is 3x higher that Red Wine's
# Volatile Acidity mean value of Red Wine is 2x higher that the White Wine's
```

```python
# Plot 8. Finding a factor 0 < k < 1, that would represent quality based on wine properties.
# The bigger the factor, the better the quality suppose to be.

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle("Quality Factor", fontsize=20, y=1.1)

k_r = (rw["citric acid"] + rw["alcohol"]) / (rw["density"] + rw["volatile acidity"])
rw["factor"] = k_r / k_r.abs().max()
axes[0].set_title("Red Wine", fontsize=15)
sns.stripplot(
    data=rw,
    x="quality label",
    y="factor",
    palette=colors,
    hue="quality label",
    ax=axes[0],
    s=5,
    jitter=0.47,
    alpha=0.6,
)
# axes[0].set(ylim=(0.45, 0.8))
axes[0].legend([], [], frameon=False)

k_w = (ww["pH"] + ww["alcohol"]) / (
    ww["density"] + ww["volatile acidity"] + ww["chlorides"]
)
ww["factor"] = k_w / k_w.abs().max()
axes[1].set_title("White Wine", fontsize=15)
sns.stripplot(
    data=ww,
    x="quality label",
    y="factor",
    palette=colors,
    hue="quality label",
    ax=axes[1],
    s=5,
    jitter=0.47,
    alpha=0.6,
)
axes[1].legend([], [], frameon=False)

k = (wine["pH"] + wine["alcohol"]) / (
    wine["density"] + wine["volatile acidity"] + wine["chlorides"]
)
wine["factor"] = k / k.abs().max()
axes[2].set_title("Wine", fontsize=15)
sns.stripplot(
    data=wine,
    x="quality label",
    y="factor",
    palette=colors,
    hue="quality label",
    ax=axes[2],
    s=5,
    jitter=0.47,
    alpha=0.6,
)
axes[2].legend([], [], frameon=False)

for ax in axes:
    for patch in ax.patches:
        patch.set_edgecolor("black")
```

```python
# Testing the probability of the factor

def k_prob(data, q):
    q_labels = ["low", "medium", "high"]
    n = q_labels.index(q)
    dql = data["quality label"]
    if n == 0:
        a = data[(data["factor"] < data[dql == q]["factor"].mean()) & (dql == q)]
    else:
        a = data[
            (data["factor"] > data[dql == q_labels[n - 1]]["factor"].mean())
            & (dql == q)
        ]
    return a.value_counts().count() / data[dql == q].value_counts().count()

print(k_prob(rw, 'low'), k_prob(rw, 'medium'), k_prob(rw, 'high'))
print(k_prob(ww, 'low'), k_prob(ww, 'medium'), k_prob(ww, 'high'))
print(k_prob(wine, 'low'), k_prob(wine, 'medium'), k_prob(wine, 'high'))
```

```python
#Plot 9. Bivariate Analysis. KDE and Scatter Plots.

def kde_scatter_plot(data, x, y):
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
    sns.kdeplot(
        data=data,
        y=y,
        x=x,
        hue="quality label",
        palette=colors,
        levels=10,
        ax=axes[0]
    )
    sns.scatterplot(
        data=data,
        y=y,
        x=x,
        hue="quality label",
        palette=colors_a,
        ax=axes[1],
        marker='s',
        s=30,
        linewidth=0
    )
    plt.show()


kde_scatter_plot(wine[wine["residual sugar"] < 20], "residual sugar", "alcohol")
kde_scatter_plot(
    rw[rw["total sulfur dioxide"] < 200], "volatile acidity", "total sulfur dioxide"
)
kde_scatter_plot(rw, "sulphates", "alcohol")
```
